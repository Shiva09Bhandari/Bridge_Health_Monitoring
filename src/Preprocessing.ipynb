## Preprocessing for Data without ChipTime

```python
date_time_str = '2020-02-18 01:30:00.0' # Sensor Start Time 
date_time_obj = datetime.datetime.strptime(date_time_str,'%Y-%m-%d %H:%M:%S.%f')
```

```python
txt_data = pd.read_csv('/home/deaddemocracy/Documents/Major_Project/Extract_Feb18_28_1PM/sensor6/sensor6_feb18_28.txt', skiprows = 1,
                       sep ='\t')
```

```python
txt_data.head()
```

```python
txt_data['rtime'] = np.arange(len(txt_data))
txt_data['rtime'] = txt_data['rtime'].map(lambda coun:(pd.Timestamp(date_time_obj) + datetime.timedelta(milliseconds = 200 * coun)))
print(txt_data.tail());
```

```python
txt_data = txt_data.drop(['Time(s)', 'ChipTime'],axis = 1)
```

```python
txt_data = txt_data.dropna(axis = 1, how = 'all')
```
